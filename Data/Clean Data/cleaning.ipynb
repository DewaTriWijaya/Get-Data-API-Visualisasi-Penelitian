```python
import pandas as pd

df_dim_authors = pd.read_csv("dim_authors.csv")

# hapus column institution
df_dim_authors = df_dim_authors.drop(columns=["institution"])
# ubah nidn menjadi integer
df_dim_authors["nidn"] = df_dim_authors["nidn"].fillna(0).astype(int)
# orcid missing valeu di isi tidak ada
df_dim_authors["orcid"] = df_dim_authors["orcid"].fillna("tidak ada")

df_dim_authors.head()

```
